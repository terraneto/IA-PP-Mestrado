```python
import itertools

# Definir os números
numeros = [1, 2, 3, 4, 5, 6, 7, 8]

# Lista para armazenar todas as combinações
combinacoes = []

# Gerar todas as combinações
for r in range(1, len(numeros) + 1):
    combinacoes.extend(itertools.combinations(numeros, r))

# Imprimir as combinações
for combinacao in combinacoes:
    print(combinacao)
```
